```python
import os 
import pandas as pd 
import matplotlib
matplotlib.rcParams.update({
    'font.family': 'serif'
})
import matplotlib.pyplot as plt
from pathlib import Path
import re
import numpy as np
import shutil
from plotly.offline import download_plotlyjs, init_notebook_mode,  iplot
pd.options.plotting.backend = "plotly"
```

```python
def plot_kinetic_energy(input_dict_l,save_name):
    fig,ax = plt.subplots(figsize=(3,3))
    for dico in input_dict_l:
        path = dico["Path"]
        name = dico["Name"]
        csv_df = pd.read_csv(path.joinpath("KineticEnergy.csv")).drop_duplicates()
        ax.plot(csv_df["timestep"],csv_df["KineticEnergy"],label=name)
    ax.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
    ax.set_xlim((0,1400))
    ax.set_ylabel("Kinetic Energy (J)")
    ax.set_xlabel("Timestep")
    plt.savefig(save_name+'.pgf',bbox_inches='tight')
```
