```python
import os 
import pandas as pd 
import matplotlib
import matplotlib.pyplot as plt
from pathlib import Path
import re
import numpy as np
import shutil
from plotly.offline import download_plotlyjs, init_notebook_mode,  iplot

pd.options.plotting.backend = "plotly"
```

```python
def plot_kinetic_energy(input_dict_l):
    fig,ax = plt.subplots()
    for dico in input_dict_l:
        path = dico["Path"]
        name = dico["Name"]
        csv_df = pd.read_csv(path.joinpath("KineticEnergy.csv")).drop_duplicates()
        ax.plot(csv_df["timestep"],csv_df["KineticEnergy"],label=name)
    ax.legend()
    ax.set_xlim((0,1500))
    ax.set_ylabel("Kinetic Energy (J)")
    ax.set_xlabel("Timestep")
    plt.show()
```
